```python
opendata_project.head(20)
```

### Missing/Null Checking

```python
opendata_project.isnull().any()#False ไม่มีค่าว่าง True มีค่าว่าง
```

####Dropna

```python
# ' price_min ' ราคาต่ำสุดโครงการ
print(len(opendata_project[opendata_project['price_min'].isnull()])) #มี row NaN 109 row 
opendata_project = opendata_project.dropna(subset=['price_min'])#we used dropna()
```

```python
# ' longitude '
print(len(opendata_project[opendata_project['longitude'].isnull()])) #มี row NaN 109 row 
opendata_project[opendata_project['longitude'].isnull()] #เรียก row ที่มี Nan ของ columns longitude
```

```python
opendata_project = opendata_project.dropna(subset=['longitude'])
opendata_project[opendata_project['longitude'].isnull()] #เรียก row ที่มี Nan ของ columns longitude
```

```python
# 'subdistrict_id' รหัสตำบล
print(len(opendata_project[opendata_project['subdistrict_id'].isnull()])) 
opendata_project = opendata_project.dropna(subset=['subdistrict_id'])
```

```python
# 'zipcode' รหัสไปรษณีย์
print(len(opendata_project[opendata_project['zipcode'].isnull()])) 
opendata_project = opendata_project.dropna(subset=['zipcode'])
opendata_project[opendata_project['zipcode'].isnull()] #เรียก row ที่มี Nan ของ columns longitude
```

####  Fill Null with real world Knownlege

```python
#fill NA with Text
opendata_project['neighborhood_id'] = opendata_project['neighborhood_id'].fillna('Unknown') #รหัสย่าน แทนค่า Na ด้วย unknow
opendata_project['neighborhood_name_th'] = opendata_project['neighborhood_name_th'].fillna('Unknown') #ชื่อย่าน แทนค่า Na ด้วย unknow
```

```python
# จำนวนลิฟต์และชั้น
opendata_project[['count_floor', 'count_unit','count_unittype','count_tower']]  = opendata_project[['count_floor', 'count_unit','count_unittype','count_tower']].fillna(1) #จำนวนชั้นของตัวโครงการ, จำนวนยูนิตในโครงการ, จำนวนแบบ้านในโครงการ , แทนค่า na ด้วย 1
opendata_project[['count_elevator','count_floor','count_parking', 'count_elevator_service']] = opendata_project[['count_elevator','count_floor','count_parking', 'count_elevator_service']].fillna(0)
# สิ่งอำนวยความสะดวก
opendata_project[['facility_clubhouse','facility_fitness','facility_meeting','facility_park','facility_playground','facility_pool','facility_security']] = opendata_project[['facility_clubhouse','facility_fitness','facility_meeting','facility_park','facility_playground','facility_pool','facility_security']].fillna(0) 
```

```python
opendata_project.isnull().any()
```
