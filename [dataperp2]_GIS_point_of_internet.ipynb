```python
def cal_seven_radius (seven_dist_list, radius):
  num = 0
  for seven in seven_dist_list :
    if seven < radius :
      num = num + 1
  return num
```

```python
record_list = list() #สร้าง list ว่าง
for id in home['post_id']: # ให้ id รัน ตาม "post_id" ทุกเลข
  thishome2 = home[home['post_id']==id]  #set ค่า
  h_lat = this_home['latitude'].values[0] #.value เอาเฉพาะค่า ไม่เอา index 
  h_lng = this_home['longitude'].values[0]  #.value เอาเฉพาะค่า ไม่เอา index 
  this_seven = seven_id[str(int(id))] ##int เอา 0 ออก -> str ตรงกัน 
  seven_dist=list() #สร้าง list ว่าง
  loc1 = [h_lat,h_lng] #กำหนดว่า loc1 = บ้าน , fix ให้บ้านหลังเดิม จนกว่า จะ loop ต่อไปเสร็จ
  for seven in this_seven:
     loc2 = [seven['lat'],seven['lng']]
     dist = geopy.distance.geodesic(loc1,loc2).m
     seven_dist.append(dist)

  record_dict ={}#ส่วนนี้ของ loop นอก สร้าง dict 
  record_dict['id'] = int(id) #เก็บค่า id จาก home['post_id']
  record_dict['d250-7/11'] = sum(np.array(seven_dist)<250) #เอาค่าที่คำนวณจาก loop ใน
  record_dict['d500-7/11'] = sum(np.array(seven_dist)<500)
  record_dict['d750-7/11'] = sum(np.array(seven_dist)<750)
  record_dict['d1000-7/11'] = sum(np.array(seven_dist)<1000)
  record_dict['d1250-7/11'] = sum(np.array(seven_dist)<1250)
  record_dict['d1500-7/11'] = sum(np.array(seven_dist)<1500)
  record_list.append(record_dict)

final_table = pd.DataFrame(record_list)
```
